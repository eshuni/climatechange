```python
import pandas as pd
from sqlalchemy import create_engine
import datetime
```

```python
df1 = pd.read_csv("https://www1.ncdc.noaa.gov/pub/orders/CDODiv4745018061866.txt")
```

```python
df1.head()
```

```python
data = df1.StateCode.replace([104], ['SouthEast'], inplace=True)
data
```

```python
temp_col_list = list(df1.columns)
temp_col_list = [item.strip() for item in temp_col_list]
print(temp_col_list)
df1.columns = temp_col_list
```

```python
data_df = df1[['StateCode', 'YearMonth', 'PCP', 'TAVG', 'PDSI', 'PHDI', 'TMIN', 'TMAX']].copy()
data_df.head()
```

```python
data_df['YearMonth'] = df1['YearMonth'].apply(lambda cell: str(cell)[0:4] + '-' + str(cell)[4:6])
data_df['YearMonth'].head()
```

```python
data_df.head()
```

```python
South_east_df = data_df.rename(columns={"YearMonth": "Date","StateCode":"Region"})
South_east_df.head(10)
```

```python
#export to csv
South_east_df.to_csv('southeast.csv', index=False)
```
