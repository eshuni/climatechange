```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "https://www1.ncdc.noaa.gov/pub/orders/CDODiv3653158064107.txt"
climate_data_df = pd.read_csv(csv_file)
climate_data_df.head()
```

```python
data = climate_data_df.StateCode.replace([101], ['Northeast'], inplace=True)
data
```

```python
temp_col_list = list(climate_data_df.columns)
temp_col_list = [item.strip() for item in temp_col_list]
print(temp_col_list)
climate_data_df.columns = temp_col_list
```

```python
ne_climate_data_df = climate_data_df[['StateCode', 'YearMonth', 'PCP', 'TAVG', 'PDSI', 'PHDI', 'TMIN', 'TMAX']].copy()
ne_climate_data_df.head()
```

```python
ne_climate_data_df['YearMonth'] = ne_climate_data_df['YearMonth'].apply(lambda cell: str(cell)[0:4] + '-' + str(cell)[4:6])
ne_climate_data_df['YearMonth'].head()
```

```python
ne_climate_data_df = ne_climate_data_df.rename(columns={"YearMonth": "Date","StateCode":"Region"})
ne_climate_data_df.head(10)
```

```python
ne_climate_data_df.to_csv('northeast.csv', index=False)
```
